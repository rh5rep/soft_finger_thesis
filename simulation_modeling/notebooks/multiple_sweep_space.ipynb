```python
import numpy as np


from simulation_modeling.model_config import routing_families, model_cases
from simulation_modeling import candidate_sweep_utils, straight_finger_sweep


JOINT_MODEL_CASES = model_cases.build_joint_space_model_cases()
STRAIGHT_MODEL_CASES = model_cases.build_straight_finger_model_cases()


candidates = routing_families.ALL_TWO_GUIDE_FAMILIES

JOINT_SWEEPS = [
    candidate_sweep_utils.SweepSpec(
        name="joint_default",
        q_sweep=candidate_sweep_utils.make_coordinated_flexion_sweep(
            mcp_start_deg=0.0,
            mcp_stop_deg=-40.0,
            n=40,
            pip_over_mcp=1.5,
            dip_over_pip=0.8,
        ),
    )
]

STRAIGHT_SWEEPS = [
    straight_finger_sweep.SweepSpec(
        name="straight_default",
        theta_sweep=list(np.deg2rad(np.linspace(0.0, -40.0, 40))),
    )
]


reject_limits = candidate_sweep_utils.RejectLimits(
    max_tension=100.0,
    max_pull=20.0,
    max_tau_error=20.0,
    max_branch_imbalance=10.0,
    min_moment_arm_norm=1e-3,
)

score_weights = candidate_sweep_utils.ScoreWeights(
    w_tension=1.0,
    w_tau_error=1.0,
    w_pull=1.0,
    w_balance=1.0,
    w_smoothness=1.0,
)
```

```python

from simulation_modeling import sweep_runner

run = sweep_runner.run_experiment(
    model_cases=JOINT_MODEL_CASES + STRAIGHT_MODEL_CASES,
    candidates=candidates,
    sweep_specs={
        "joint_space": JOINT_SWEEPS,
        "straight_finger": STRAIGHT_SWEEPS,
    },
    reject_limits=reject_limits,
    score_weights=score_weights,
)
```

```python
raw_path, summary_path, raw_history_path, summary_history_path = sweep_runner.export_candidate_results(
    outputs=run.outputs,
    results=run.results,
)
raw_path, summary_path
```
